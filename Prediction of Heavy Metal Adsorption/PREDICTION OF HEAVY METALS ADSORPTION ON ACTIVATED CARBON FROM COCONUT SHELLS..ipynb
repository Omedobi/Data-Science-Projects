
Adsorption capacity Qe 

`Qe=(C0-Ce)×V/m`

Where:

V = The volume of the solution\
m = The mass of adsorbent (g)\
The percentage of metal ions adsorbed from solutions was calculated by the following equation;\

`%adsorbed =(C0-Ce)/C0×100`

C0=Initial concentration of heavy meal\
Ce=Final concentration of heavy metal

Langmuir Isotherm: The Langmuir isotherm is expressed as:

`Ce/Qe=Ce(1/Q¬m) +1/bQm`

Where:

Qe= The equilibrium value of the metal ions adsorbed per unit weight of adsorbent\
Ce= The equilibrium concentration of the adsorbate\
b= Relates the affinity between the adsorbent and the adsorbate\
The favourability of the isotherm depends on the dimensionless constant factor R1 which is given by:

`RL=1/1+KL*C0`

C0=Initial concentration 
If RL=0, it shows that the adsorption isotherm is irreversible and if 0<RL<1 then it shows that Langmuir adsorption isotherm is favourable.

Freundlich Isotherm: The Freundlich isotherm was expressed by the equation:

`logQe=logkf+1/nlogCe`

Where:

kf=The adsorption coefficient which characterize the strength of adsorption\
Ce=the equilibrium concentration of the adsorbate\
Qe=the equilibrium value of the adsorbed per unit weight

**EDA on the parameters**

-The exploratory data analysis (EDA) plays a crucial role in understanding the relationships and patterns within datasets involving parameters such as pH, temperature, adsorption, and time, along with contaminants like lead, zinc, and cadmium. Here are some types of EDA that can be performed:

1. **Descriptive Statistics**: Compute basic statistics such as mean, median, mode, standard deviation, and range for each parameter and contaminant. This provides a summary of the central tendency and variability of the data.

2. **Data Visualization**:
   - **Scatter Plots**: Plotting the relationship between variables such as pH, temperature, and adsorption capacity can help identify any correlations or trends.
   - **Histograms**: Visualize the distribution of each parameter and contaminant concentration to understand their frequency distribution.
   - **Box Plots**: Identify the spread of data and any potential outliers in each parameter and contaminant group.
   - **Time Series Plots**: If time is a factor, plotting the variation of parameters and contaminant concentrations over time can reveal temporal patterns.

3. **Modeling and Prediction**:
   - Develop predictive models to estimate adsorption efficiency based on pH, temperature, and time using techniques such as regression analysis or machine learning algorithms.
   - Validate the models using techniques like cross-validation and assess their accuracy and reliability.


-By conducting these types of EDA, researchers in the environmental technology field can gain valuable insights into the adsorption process and optimize the conditions for efficient removal of heavy metals from contaminated water sources.

Histogram

```python
def plot_histograms_subplot(datasets, titles, columns, colors):
    num_datasets = len(datasets)
    num_cols = len(columns)
    num_rows = (num_cols + 1) // 2  

    fig, axes = plt.subplots(nrows=num_datasets, ncols=num_cols, figsize=(16, 8*num_datasets))
    axes = axes.flatten()

    for i, data in enumerate(datasets):
        for j, (column, color) in enumerate(zip(columns, colors)):
            ax = axes[i * num_cols + j]
            ax.hist(data[column], bins=20, alpha=0.5, label=column, color=color)
            ax.set_title(f'{titles[i]} ({column}, {data.name})')  
            ax.set_xlabel('Concentration (Ce)')
            ax.set_ylabel('Frequency')
            ax.legend()

    for k in range(num_datasets * num_cols, num_datasets * num_rows):
        axes[k].axis('off')

    fig.suptitle('Histograms of Concentration (Ce) for different contaminant and parameters', fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
datasets = [time_df, temp_df, ads_df, pH_df] 
titles = ['Concentration (Ce)', 'Concentration (Ce)', 'Concentration (Ce)', 'Concentration (Ce)']  
columns = ['Zinc_Ce', 'Lead_Ce', 'Cadmium_Ce']  
colors = ['tab:red', 'tab:blue', 'tab:green']  

time_df.name = 'time'
temp_df.name = 'temperature'
ads_df.name = 'adsorption'
pH_df.name = 'pH'

plot_histograms_subplot(datasets, titles, columns, colors)
```

Line Plot

```python
def plot_lineplots_subplot(datasets, titles, parameters, colors):
    num_datasets = len(datasets)
    num_params = len(parameters[0])

    fig, axes = plt.subplots(nrows=num_datasets, ncols=num_params, figsize=(18, 8*num_datasets))
    axes = axes.flatten()

    for i, (data, title) in enumerate(zip(datasets, titles)):
        for j, ax in enumerate(axes[i*num_params:(i+1)*num_params]):
            for param, color in zip(parameters[i], colors):
                ax.plot(data.index, data[param], label=param.split('_')[0], color=color)
            ax.set_title(f'{title} ({data.name})')
            ax.set_xlabel(f'{data.name}')
            ax.set_ylabel('Quantity (g)')
            ax.legend()

    fig.suptitle('Line Plots of Parameters for Different Datasets', fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
datasets = [time_df, temp_df, ads_df, pH_df]
titles = ['Adsorption (%)', 'Adsorption (%)', 'Adsorption (%)', 'Adsorption (%)']  
parameters = [['Zinc_% adsorbed', 'Cadmium_% adsorbed', 'Lead_% adsorbed'], 
              ['Zinc_% adsorbed', 'Cadmium_% adsorbed', 'Lead_% adsorbed'],
              ['Zinc_% adsorbed', 'Cadmium_% adsorbed', 'Lead_% adsorbed'],
              ['Zinc_% adsorbed', 'Cadmium_% adsorbed', 'Lead_% adsorbed']]  
colors = ['tab:red', 'tab:blue', 'tab:green']

time_df.name = 'time'
temp_df.name = 'temperature'
ads_df.name = 'adsorption'
pH_df.name = 'pH'

plot_lineplots_subplot(datasets, titles, parameters, colors)
```

Boxplot

```python
def plot_boxplots_subplot(datasets, titles):
    num_datasets = len(datasets)

    fig, axes = plt.subplots(nrows=num_datasets, figsize=(12, 6*num_datasets))

    for i, (data, title) in enumerate(zip(datasets, titles)):
        ax = axes[i] if num_datasets > 1 else axes
        sns.boxplot(data=data, orient='h', ax=ax)
        ax.set_title(title)

    fig.suptitle('Boxplots of Numerical Variables for Different Datasets', fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

**Remove Outliers from the train data using Z_score**

```python
from scipy import stats

def remove_outliers(X, y, threshold=3):
    selected_columns = X.columns
    z_scores = np.abs(stats.zscore(X[selected_columns]))
    outlier_indices = np.where(z_scores > threshold)[0]
    X = X.drop(X.index[outlier_indices])
    y = y.drop(y.index[outlier_indices])
    return X, y

X_temp_train, y_temp_train = remove_outliers(X_temp_train, y_temp_train)

X_time_train, y_time_train = remove_outliers(X_time_train, y_time_train)

X_ads_train, y_ads_train = remove_outliers(X_ads_train, y_ads_train)

X_pH_train, y_pH_train = remove_outliers(X_pH_train, y_pH_train)

print("temp_df Train set after removing outliers:", X_temp_train.shape, y_temp_train.shape)
print("time_df Train set after removing outliers:", X_time_train.shape, y_time_train.shape)
print("ads_df Train set after removing outliers:", X_ads_train.shape, y_ads_train.shape)
print("pH_df Train set after removing outliers:", X_pH_train.shape, y_pH_train.shape)
```
